```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing the dataset
df_train = pd.read_csv('data/train.csv')
df_train.head()
```

```python
df_train.info()
```

```python
#importing the test dataset
df_test = pd.read_csv('data/test.csv')
df_test.head()
```

```python
#Merging both datasets
df = pd.concat([df_train, df_test], sort=False)
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["User_ID"] , axis=1, inplace=True)
```

```python
df.head()
```

```python
df["Gender"] = df["Gender"].map({"F":0, "M":1})
```

```python
df.head()
```

```python
df['Age'].unique()
```

```python
## Handling categorical data (Featuring Age)

df["Age"] = df["Age"].map({"0-17":1, "18-25":2, "26-35":3, "36-45":4, "46-50":5, "51-55":6, "55+":7})
```

```python
from sklearn import preprocessing
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'species'.
df['Age']= label_encoder.fit_transform(df['Age'])
 
df['Age'].unique()
```

```python
df.head()
```

**Fixing categorical Category**

## Observations (Key points to keep in Mind) :-
 As we can see in the above output there are a lot of Null Values in the Product Categories . So what should we do now ? 
 - The answer is we basically will fix it by replacing the values
    - Why are we replacing it ?
    - Ans :- Beacause we are ging to train a model and it will be having some functions and it we give it this much null values it will not work perfectly. 

```python
# Replacing NaN values in Product_Category_2 and Product_Category_3 with 0
# df['Product_Category_2'].unique()
df['Product_Category_2'].value_counts()
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

### Why we did not replace the Null values by 0 ?
#### Because we can not provide 0 as a value to the model while training otherwise it will give false results 

```python
df["Product_Category_2"].isnull().sum()
```

```python
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```
